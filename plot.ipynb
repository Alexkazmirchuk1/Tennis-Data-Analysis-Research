```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style='whitegrid', palette='colorblind')

# Calculate the probability of holding serve
def prob_hold(p):
    """Probability server holds."""
    q = 1 - p
    return (
        p**4 +
        4 * (p**4) * q + 
        10 * (p**4) * (q**2) + 
        (20 * (p**3) * (q**3)) * ((p**2) / (1 - 2 * p * q))
    )

# Calculate the probability of winning a game given the probability of winning a point
def prob_win_game(p):
    """Probability of winning a game given point winning probability."""
    q = 1 - p
    return (
        p**4 +                          # Player wins the game on first four points
        4 * (p**4) * q +                # Player wins the game with one point loss
        10 * (p**4) * (q**2) +          # Player wins the game with two points lost
        20 * (p**3) * (q**3) * (p**2) / (1 - 2 * p * q)  # Player wins after losing three points
    )

# Calculate the probability of winning a set given the probability of winning a game
def prob_win_set(ps):
    """Probability of winning a set."""
    qs = 1 - ps
    return (
        ps**6 +
        6 * (ps**6) * qs + 
        21 * (ps**6) * (qs**2) + 
        56 * (ps**6) * (qs**3) +
        126 * (ps**6) * (qs**4) +
        42 * (ps**7) * (qs**5) +
        924 * (ps**7) * (qs**6)
    )

# Calculate the probability of winning a match given the probability of winning a set
def prob_win_match(pm):
    """Probability of winning a match."""
    qm = 1 - pm
    return (
        pm**3 +
        3 * (pm**3) * qm + 
        6 * (pm**3) * (qm**2)
    )

# Generate an array of probabilities of winning a point
point_win_probs = np.linspace(0, 1, 100)
game_probs = []
match_probs = []

# Calculate the probabilities for games and matches for each point winning probability
for p in point_win_probs:
    if p == 0:  # Handle the case where the player cannot win any points
        game_probability = 0
        match_probability = 0
    elif p == 1:  # If the player wins every point
        game_probability = 1
        match_probability = 1
    else:
        game_probability = prob_win_game(p)
        set_probability = prob_win_set(game_probability)
        match_probability = prob_win_match(set_probability)

    game_probs.append(game_probability)
    match_probs.append(match_probability)

# Plotting the Probability of Winning a Game
plt.figure(figsize=(10, 6))
plt.plot(point_win_probs, game_probs, label='Probability of Winning Game', color='orange', lw=2)
plt.title('Probability of Winning a Game as a Function of Winning Each Point', fontsize=16)
plt.xlabel('Probability of Winning Each Point', fontsize=14)
plt.ylabel('Probability of Winning Game', fontsize=14)
plt.grid()
plt.legend(fontsize=12)
plt.ylim(0, 1)
plt.xlim(0, 1)
plt.axhline(0, color='grey', lw=0.5, ls='--')
plt.axvline(0, color='grey', lw=0.5, ls='--')
plt.show()

# Plotting the Probability of Winning a Match
plt.figure(figsize=(10, 6))
plt.plot(point_win_probs, match_probs, label='Probability of Winning Match', color='blue', lw=2)
plt.title('Probability of Winning a Match as a Function of Winning Each Point', fontsize=16)
plt.xlabel('Probability of Winning Each Point', fontsize=14)
plt.ylabel('Probability of Winning Match', fontsize=14)
plt.grid()
plt.legend(fontsize=12)
plt.ylim(0, 1)
plt.xlim(0, 1)
plt.axhline(0, color='grey', lw=0.5, ls='--')
plt.axvline(0, color='grey', lw=0.5, ls='--')
plt.show()
```
